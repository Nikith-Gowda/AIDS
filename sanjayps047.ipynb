```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/spam_ham_emails.csv")
```

```python
df.shape
```

```python
for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].str.lower()
```

```python
df
```

```python
import tensorflow as tf
from tensorflow import keras

keras.datasets.mnist.load_data()

(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()

X_train.shape
len(X_train)

X_train_flattened = X_train.reshape(len(X_train),28*28)
X_test_flattened = X_test.reshape(len(X_test),28*28)

model = keras.Sequential([keras.layers.Dense(10,input_shape = (784,),activation='sigmoid')])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(X_train_flattened, y_train,epochs=5)

y_test_predicted = model.predict(X_test_flattened)

print(y_test_predicted[2])

y_test_predicted = model.predict(X_test_flattened)

y_test_predicted.shape

np.argmax(y_test_predicted[2])

y_predict_labels = [np.argmax(i) for i in y_test_predicted]

y_predict_labels[:5]

y_test[:5]

cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predict_labels)

plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')

model = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)), keras.layers.Dense(100, activation='relu'), keras.layers.Dense(10, activation='sigmoid')])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(X_train, y_train,epochs=50)

model.evaluate(X_test,y_test)

y_predicted = model.predict(X_test)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predict_labels)

plt.figure(figsize = (7,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')

plt.matshow(X_test[78])
plt.matshow(X_train[78])

y_test[78]
y_train[78]
```
